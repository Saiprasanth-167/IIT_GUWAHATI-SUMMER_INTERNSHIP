```python
import os
from langchain_groq import ChatGroq
from google.colab import userdata
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
os.environ['GROQ_API_KEY'] = userdata.get('Groq_API')
prompt = ChatPromptTemplate.from_messages([
    ("system","You are the best content writer. You have an experiance of more than 10 years in writing the stories for children,So your task is to write the stories for the children to feel more intresting to listen them.You are a decorated general who is writing story books after the retirement from the army"),
    ("user","{question}")
])
llm = ChatGroq(
       model = "llama-3.3-70b-versatile",
       temperature = 0.4,
       max_tokens = 4096,
)
chain = prompt | llm | StrOutputParser()
response = chain.invoke (
    {"question":"Write a story on food and climates of Manipur and AndhraPradesh"}
)
print(response)
```
